```python
#histogram
plt.figure(figsize=(8,6))
sns.histplot(df['sepal.length'])
plt.xlabel("sepallength(cm)")
plt.title("histogram of sepal length")
plt.show()
```

```python
#boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x="variety",y="sepal.length",data=df)
plt.xlabel("species")
plt.ylabel("sepal length(cm)")
plt.title("box plot")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(df["sepal.width"],fill=True)
plt.xlabel("Sepal Width(cm)")
plt.ylabel("Density")
plt.title("Distribution chart")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="sepal.length",y="sepal.width",hue="variety",data=df)
plt.ylabel("Sepal Width(cm)")
plt.xlabel("Sepal length(cm)")
plt.title("scatterplot")
plt.show()
```

```python
#scatterplot
plt.figure(figsize=(8,6))
sns.scatterplot(x="petal.length",y="petal.width",hue="variety",data=df)
plt.ylabel("petal Width(cm)")
plt.xlabel("petal length(cm)")
plt.title("scatterplot")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="petal.length",y="petal.width",hue="variety",data=df,size="sepal.length")
plt.ylabel("petal Width(cm)")
plt.xlabel("petal length(cm)")
plt.title("scatterplot")
plt.show()
```

```python
#scatter matrix(pair plot)
plt.figure(figsize=(8,6))
sns.pairplot(df,hue="variety",markers=("o","s","d"))
plt.suptitle("scatterMatrix (pair plot)",y=1.02)
plt.show()
```

```python
#bubble chart
plt.figure(figsize=(8,6))
sns.scatterplot(x="sepal.length",y="sepal.width",hue="variety",data=df,size="petal.length")
plt.ylabel("sepal Width(cm)")
plt.xlabel("sepal length(cm)")
plt.title("bubble chart")
plt.show()
```

```python
#density
plt.figure(figsize=(8,6))
sns.kdeplot(x=df["sepal.length"],y=df["sepal.width"],cmap="Blues",fill=True)
plt.ylabel("sepal Width(cm)")
plt.xlabel("sepal length(cm)")
plt.title("Density")
plt.show()
```

```python
#parallel coordinate chart
plt.figure(figsize=(8,6))
sns.set(style="whitegrid")
pd.plotting.parallel_coordinates(df,"variety",colormap="viridis")
plt.title("parallel coordinates chart")
plt.show()
```

```python
#deviation chart
from matplotlib import markers
from seaborn._core.properties import LineStyle
features="sepal.length"
meanvalue=df[features].mean()
df["deviation"]=df[features]-meanvalue
plt.figure(figsize=(8,6))
plt.plot(df.index,df["deviation"],marker='s',linestyle="-",color='y')
plt.xlabel("datapoints")
plt.ylabel(f"deviation form mean {features.capitalize()}")
plt.title(f"deviation chart for{features.capitalize()} ")
plt.grid(True)
plt.show()
```

```python
#andrews curve
plt.figure(figsize=(8,6))
sns.set(style="ticks")
pd.plotting.andrews_curves(iris,"species",colormap="viridis")
plt.title("andrews curve")
plt.show()
```
