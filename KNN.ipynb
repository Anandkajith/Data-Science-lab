```python
#libries
import numpy as np
import pandas as pd
```

```python
#dataset implementation
irisData=pd.read_csv('/content/iris.csv')
```

```python
print(irisData.describe())
```

```python
#count of classes types
irisData.groupby('variety').size()
```

```python
#features and target divide
features=['sepal.length','sepal.width','petal.length','petal.width']
X=irisData[features].values
Y=irisData['variety'].values
```

```python
#label converstion of target
# 0 - Setosa 1- Versicolor 2-Virginica
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
Y=le.fit_transform(Y)
```

```python
print(Y)
```

```python
#train and test datasplit
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size = 0.2, random_state=42)
```

```python
#import Knn files
from sklearn.neighbors import KNeighborsClassifier
```

```python
#fitting of KNN
classifier=KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train,Y_train)
```

```python
#testing of data
print(classifier.predict(X_test))
```

```python
print(Y_test)
```

```python
a=[[4.1,2.2,3.2,0.1]]
if(classifier.predict(a)==0):
  print("Setosa")
elif (classifier.predict(a)==1):
  print("Versicolor")
else:
  print("Virginica")
```
